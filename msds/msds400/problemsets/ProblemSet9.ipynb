```python
import networkx as nx
```

```python

def draw_graph(G):
    pos=nx.spring_layout(G) # pos = nx.nx_agraph.graphviz_layout(G)
    nx.draw_networkx(G,pos)
    labels = nx.get_edge_attributes(G,'weight')
    nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
```

```python
G=nx.Graph()
G.add_edge('a', 'b', weight=4)
G.add_edge('a', 'c', weight=3)

G.add_edge('b', 'd', weight=5)
G.add_edge('b', 'c', weight=2)

G.add_edge('c', 'd', weight=3)
G.add_edge('c', 'e', weight=6)

G.add_edge('d', 'f', weight=5)
G.add_edge('d', 'e', weight=1)

G.add_edge('e', 'g', weight=5)


G.add_edge('f', 'g', weight=2)
G.add_edge('f', 'z', weight=7)

G.add_edge('g', 'z', weight=4)

draw_graph(G)
```

```python
def lowest_cost(G, v, u):
    shortest_path = nx.dijkstra_path(G, v, u)
    cost = sum(list(G[i][j]['weight'] for i, j in zip(shortest_path[1:], shortest_path[:-1])))
    return cost, shortest_path
```

```python
lowest_cost(G, 'a', 'z')
```

```python
G=nx.Graph()
G.add_edge('a', 'c')
G.add_edge('a', 'b')
G.add_edge('a', 'e')

G.add_edge('b', 'e')
G.add_edge('b', 'c')

G.add_edge('c', 'd')

G.add_edge('d', 'e')

draw_graph(G)
```

```python
print(f"is_eulerian: {nx.is_eulerian(G)}")
print(f"has_eulerian_path: {nx.has_eulerian_path(G)}")
```

```python
G=nx.Graph()
G.add_edges_from([
    ('a', 'b'),
    ('a', 'e'),
    ('a', 'c'),
    ('b', 'e'),
    ('b', 'e'),

])
draw_graph(G)
print(f"is_eulerian: {nx.is_eulerian(G)}")
print(f"has_eulerian_path: {nx.has_eulerian_path(G)}")
```
